```python
import torch
from torch.autograd import Variable
```

```python
X = [1,2,3,4,5]
y  = [2,4,6,8,10]

w = Variable(torch.Tensor([1.0]), requires_grad = True)
```

```python
def forward(x):
     return x*w
    
    
def loss(x,y):
    y_predict = forward(x)
    return ((y_predict - y)*(y_predict - y))
 
```

```python
for epoch in range(9):
    for x_val,y_val in zip(X,y):
        l = loss(x_val,y_val)
        l.backward()
        w.data = w.data - 0.01*w.grad.data
        w.grad.data.zero_()
    print("epoch and loss",epoch,l.data[0])
```

```python
import torch 
from torch.autograd import Variable

X = Variable(torch.Tensor([[1.0],[2.0],[3.0]]))
y = Variable(torch.Tensor([[2.0],[4.0],[6.0]]))

```

```python
class model(torch.nn.Module):
    
    def __init__(self):
        super(model,self).__init__()
        self.liner = torch.nn.Linear(1,1)  # para are in,out  (1,1)
        
    def forward(self,x):
        y_predict = self.liner(x)
        return y_predict
    
Model = model()


```

```python
loss = torch.nn.MSELoss(size_average=False)
optimizer = torch.optim.SGD(Model.parameters(),lr = 0.01)

```

```python
for epoch in range(100):
    
    y_predict = Model(X)
    l = loss(y_predict,y)
    print(epoch,l.data[0])
    optimizer.zero_grad()
    l.backward()
    optimizer.step()
    
```

```python
from torch.nn import functional as F
X = Variable(torch.Tensor([[1.0],[2.0],[3.0],[4.0]]))
y  = Variable(torch.Tensor([[0.],[1.],[0.],[1.]]))

class logi(torch.nn.Module):
    def __init__(self):
        super(logi, self).__init__()
        self.linear = torch.nn.Linear(1,1)
        
    def forward(self,x):
        y_predict = F.sigmoid(self.linear(x))
        return y_predict
    
model = logi()

loss = torch.nn.BCELoss(size_average=True)
optimizer = torch.optim.SGD(model.parameters(),lr=0.01)
for epoch in range(500):
    
    y_pre  = model(X)
    
    l = loss(y_pre,y)
    print(epoch,l.data[0])
    optimizer.zero_grad()
    l.backward()
    optimizer.step()
    

    
```

```python
class d(torch.nn.Module):
    def __init__(self):
        super(d,self).__init__()
        self.l1 = torch.nn.Linear(8,6)
        self.l2 = torch.nn.Linear(6,4)
        self.l3 = torch.nn.Linear(4,1)
        
        self.sigmoid = torch.nn.Sigmoid()
        
    def forward(self,x):
        
        out1 = self.sigmoid(self.l1(x))
        out2 = self.sigmoid(self.l2(out1))
        out3 = self.sigmoid(self.l3(out2))
        return out3
    
model = d()

loss = torch.nn.BCELoss(size_average=True)
optimizer = torch.optim.SGD(model.parameters(),lr = 0.1)
for epoch in range(100):
    predict = model(X)
    l = loss(predict,y)
    print(epoch,l.data[0])
    optimizer.zero_grad()
    l.backward()
    optimizer.step()
    
```

```python
import torch 
from torch.autograd import Variable
from torch.utils.data import Dataset,DataLoader
```

```python
### diabetes prediction model#####
class d(torch.nn.Module):
    def __init__(self):
        super(d,self).__init__()
        self.l1 = torch.nn.Linear(8,6)
        self.l2 = torch.nn.Linear(6,4)
        self.l3 = torch.nn.Linear(4,1)
        
        self.sigmoid = torch.nn.Sigmoid()
        
    def forward(self,x):
        
        out1 = self.sigmoid(self.l1(x))
        out2 = self.sigmoid(self.l2(out1))
        out3 = self.sigmoid(self.l3(out2))
        return out3
    
model = d()

loss = torch.nn.BCELoss(size_average=True)
optimizer = torch.optim.SGD(model.parameters(),lr = 0.1)
```

```python
##### diabetes data loader class

class dd(Dataset):
    def __init__(self):
        xy = np.loadtxt('./data/diabetes.csv.gz', delimiter=',', dtype=np.float32)
        self.len = xy.shape[0]
        self.X = torch.from_numpy(xy[:,0:-1])
        self.y = torch.from_numpy(xy[:,[-1]])
        
    def __getitem__(self,index):
        return self.X[index] ,self.y[index]
    
    def __len__(self):
        return self.len
    
dataset = dd()

train_loader = DataLoader(dataset = dataset,batch_size=32,shuffle=True)
```

```python
for epoch in range(50):
    for i,data in enumerate(train_loader,0):
        X,y = data
        X,y = Variable(X),Variable(y)
        predict = model(X)
        l = loss(predict,y)
        print(epoch,l.data[0])
        optimizer.zero_grad()
        l.backward()
        optimizer.step()
```
